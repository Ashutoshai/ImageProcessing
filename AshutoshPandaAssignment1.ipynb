```
import numpy as np
from urllib.request import urlopen
import cv2
from matplotlib import pyplot as plt
```

```
# METHOD #1: OpenCV, NumPy, and urllib
def url_to_image(url):
	# download the image, convert it to a NumPy array, and then read
	# it into OpenCV format
	resp = urlopen(url)
	image = np.asarray(bytearray(resp.read()), dtype="uint8")
	image = cv2.imdecode(image, cv2.IMREAD_COLOR)
 
	# return the image
	return image
```

```
image = url_to_image("https://fontsarena-cd5e.kxcdn.com/wp-content/uploads/2019/04/helvetica-now-font-400x364.png")
from google.colab.patches import cv2_imshow
cv2_imshow(image)
```

```
edges = cv2.Canny(image,100,200)
#cv2_imshow(edges)
#kernel = np.float32([[-1,0,1],[-1,0,1],[-1,0,1]])
#dst = cv2.filter2D(image,-1,kernel)
#cv2_imshow(dst)
kernelHorizontal = np.float32([[-1,-1,-1],[2,2,2],[-1,-1,-1]])
dstHorizontal = cv2.filter2D(image,-1,kernelHorizontal)
cv2_imshow(dstHorizontal)
```

```
#45 Degree Angle Detector (either)
kernel45Degree = np.float32([[-2,-1,0],[-1,0,1],[0,1,2]])
dst45Degree = cv2.filter2D(image,-1,kernel45Degree)
cv2_imshow(dst45Degree)
```

```
kernel135Degree = np.float32([[2,-2,-2],[-2,2,-2],[-2,-2,2]])
dst135Degree = cv2.filter2D(image,-1,kernel135Degree)
cv2_imshow(dst135Degree)
```

```
#Blur Kernel
kernelBlur = np.float32([[1/9,1/9,1/9],[1/9,1/9,1/9],[1/9,1/9,1/9]])
dstBlur = cv2.filter2D(image,-1,kernelBlur)
cv2_imshow(dstBlur)
```

```
#Sharpen Kernel
kernelSharpen = np.float32([[0,-1,0],[-1,5,-1],[0,-1,0]])
dstSharpen = cv2.filter2D(image,-1,kernelSharpen)
cv2_imshow(dstSharpen)
```

```
#Identity function
kernelIdentity = np.float32([[0,0,0],[0,1,0],[0,0,0]])
dstIdentity = cv2.filter2D(image,-1,kernelIdentity)
cv2_imshow(dstIdentity)
```
